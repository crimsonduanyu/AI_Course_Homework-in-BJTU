### 编程实现谱聚类算法对鸢尾花数据集进行聚类分析，需画图比较不同聚类簇数对结果的影响。

```python
import sklearn
from sklearn.datasets import load_iris

data = load_iris()
X = data['data']
X
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:, 0]+X[:, 2], X[:, 1]+X[:, 3], marker='o')
plt.show()
```

```python
from sklearn.cluster import SpectralClustering
from sklearn import metrics
```

##### 寻找最优参数,这个块不建议运行第二次，太恶心了，结果已经得到了，是gamma=0.2，n_cluster=3

```python
for i in range(2,10):
    y_pred = SpectralClustering(n_clusters=i, gamma=0.2).fit_predict(X)
    score = metrics.calinski_harabasz_score(X, y_pred)

    import matplotlib.pyplot as plt

    plt.subplot(2,4,i-1)
    plt.scatter(X[:, 0]+X[:, 2], X[:, 1]+X[:, 3], marker='o', c=y_pred)
    plt.title('n_cluster={}'.format(i))

#import matplotlib

#matplotlib.rcParams['figure.figsize']=(600,300)
plt.suptitle("Figures of different n_cluster values")
plt.show()
```

### Best Model

```python
# Best parameter is: [gamma:0.2,n_cluster:3]
y_pred = SpectralClustering(n_clusters=3, gamma=0.2).fit_predict(X)
score = metrics.calinski_harabasz_score(X, y_pred)
print(y_pred)

import matplotlib.pyplot as plt

plt.scatter(X[:, 0]+X[:, 2], X[:, 1]+X[:, 3], marker='o', c=y_pred)
print("The Prediction of best-param model is:")
plt.show()
```
