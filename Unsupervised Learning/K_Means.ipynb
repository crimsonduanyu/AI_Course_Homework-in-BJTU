### 编程实现k-均值聚类算法对sklearn库中的鸢尾花数据集进行聚类分析，需画图比较不同聚类簇数对结果的影响。

```python
import sklearn
from sklearn.datasets import load_iris

data = load_iris()
X = data['data']
X
```

### IRIS数据集

```python
from sklearn.cluster import KMeans

iris_model = KMeans(n_clusters=3,
                     max_iter=10000,
                     init='k-means++')

iris_model.fit(X)
predict = iris_model.predict(X)
print(predict)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:, 0]+X[:, 2], X[:, 1]+X[:, 3], marker='o', c=predict)
print("The Prediction of this model is:")
plt.show()
```

```python
from sklearn import metrics

print('cluster_centres:\n',iris_model.cluster_centers_)
#聚类中心
print('inertia:\n',iris_model.inertia_)
#所有样本点到其最近点距离之和
print('silhouette_score:\n',metrics.silhouette_score(X, predict))
#样本轮廓系数
```

### BuddyMove数据集

```python
import pandas as pd
b_data = pd.read_csv('buddymove.csv')
dataframe = b_data.loc[:,['Sports' , 'Religious' , 'Nature' , 'Theatre' , 'Shopping' , 'Picnic']]
b_data = dataframe.values
print(b_data)
```

```python
for i in range(2,10):
    bdm_model = KMeans(n_clusters=i,
                         max_iter=10000,
                         init='k-means++')

    bdm_model.fit(b_data)
    b_predict = bdm_model.predict(b_data)
    #print(b_predict)
    plt.scatter(x=b_data[:,0]+b_data[:,2]+b_data[:,4],
                y=b_data[:,1]+b_data[:,3]+b_data[:,5],
                c=b_predict)
    print("While n={}, The Prediction of this model is:".format(i))
    plt.show()
    print('cluster_centres:\n',bdm_model.cluster_centers_)
    #聚类中心
    print('inertia:\n',bdm_model.inertia_)
    #所有样本点到其最近点距离之和
    print('silhouette_score:\n',metrics.silhouette_score(b_data, b_predict))
    #样本轮廓系数
```
