### 编程实现密度聚类算法对鸢尾花数据集进行聚类分析，需画图比较不同邻域参数（ε, MinPts）对结果的影响。

```python
from sklearn.datasets import load_iris
from sklearn.cluster import DBSCAN
from sklearn import metrics
```

```python
data = load_iris()
X = data['data']
X
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:, 0]+X[:, 2], X[:, 1]+X[:, 3], marker='o')
plt.show()
```

```python
for i in range(2,7):
    for j,epsilon in enumerate((0.2, 0.5, 0.82, 1)):
        iris_model = DBSCAN(eps=epsilon,
                            min_samples=i,
                            metric='euclidean',
                            algorithm='auto')
        y_pred = iris_model.fit_predict(X)
        score = metrics.calinski_harabasz_score(X, y_pred)
        #print(y_pred)

        import matplotlib.pyplot as plt

        plt.subplot(5,4,i-1+j*5)
        plt.scatter(X[:, 0]+X[:, 2], X[:, 1]+X[:, 3], marker='o', c=y_pred)
        plt.title('eps={},minpt={}'.format(epsilon,i))

plt.suptitle("Figures of different eps & minpt values")
plt.show()
```

#### Best parameter is: [Epsilon:0.82,MinPt:2]

```python
iris_model = DBSCAN(eps=0.82,
                    min_samples=2,
                    metric='euclidean',
                    algorithm='auto')
y_pred = iris_model.fit_predict(X)
score = metrics.calinski_harabasz_score(X, y_pred)
print(y_pred)

import matplotlib.pyplot as plt

plt.scatter(X[:, 0]+X[:, 2], X[:, 1]+X[:, 3], marker='o', c=y_pred)
print("The Prediction of best-param model is:")
plt.show()
```
